```python
import pandas as pd
```

```python
df = pd.read_csv("resultado.csv")
df.head()
```

```python
df["habilidades_ordem"] = df["habilidades_ordem"].str.replace(" | ",",", regex=False).str.split(",")
df["habilidades_operador"] = df["habilidades_operador"].str.replace(" | ",",", regex=False).str.split(",")

def comparar_listas(lista1, lista2):
    # Diferença simétrica: elementos que estão em uma lista mas não na outra
    return list(set(lista1) - set(lista2))

def comparar_tamanho_listas(lista1, lista2):
    # Diferença simétrica: elementos que estão em uma lista mas não na outra
    diff = list(set(lista1) - set(lista2))
    return len(diff)/len(lista1)

# Aplicando a função de comparação nas colunas
df['diferenca_habilidades'] = df.apply(lambda row: comparar_listas(row['habilidades_ordem'], row['habilidades_operador']), axis=1)
df['habilidades_atendidas'] = df.apply(lambda row: comparar_tamanho_listas(row['habilidades_ordem'], row['habilidades_operador']), axis=1)
```
